```python
import pandas as pd
import unicodedata
```

```python
ch_1=pd.read_csv('../data/input/IMDB-Movie-Data.csv')
ch_1
```

```python
def votes(x):
    if 0 <= x <= 999:
        return 1
    elif 1000<=x<=9999:
        return 2
    elif 10000<=x<=99999:
        return 3
    elif 100000<=x<=999999:
        return 4
    elif x>=1000000:
        return 5
    
```

```python
ch_1['Bin'] = ch_1['Votes'].apply(votes)
ch_1.head(1)
```

```python
ch_1['Bin'] = ch_1.apply(lambda angel : votes(angel['Votes']), axis=1)
ch_1.head(1)
```

```python
def new_rk(rk,gen):
    if 'Thriller' in gen:
        return rk + 1
    elif 'Comedy' in gen:
        return rk - 1
    else:
        return rk
```
